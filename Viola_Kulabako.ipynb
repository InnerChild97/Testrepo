#The Wallet Wizard
##A mini FinTech budgeting story in Python
###The aim is to build a simple budgeting assistant that helps me understand where my money goes and how to improve it.

## Scene 1. The Problem
Every month, my money disappears like magic

![Money image](https://images.unsplash.com/photo-1554224155-6726b3ff858f?q=80&w=800&auto=format&fit=crop)

Want a quick budgeting method? Try the **50/30/20 rule**:  
 https://www.investopedia.com/ask/answers/022916/what-502030-budget-rule.asp


```python
# Scene 2. Adress my income + categories
monthly_income = 2200
categories = ["Rent", "Food", "Transport", "Bills", "Savings", "Michellaneous"]
```

## Scene 3. My spending diary  
I tracked my spending for a week and estimated my monthly totals.

Next, I’ll store the data in a Python dictionary.


```python
spending = {"Rent": 900,
    "Food": 320,
    "Transport": 140,
    "Bills": 180,
    "Savings": 200,
    "michellaneous": 260}
spending
```

```python
total_spent = sum(spending.values())
leftover = monthly_income - total_spent

print("Total spent:", total_spent)
print("Money left:", leftover)
```

## Scene 4: Is this budget healthy?  
I’ll use a conditional statement:

- If money left is **positive**, I’m okay.
- If money left is **negative**, I’m overspending.


```python
if leftover > 0:
    print("Good news! I have money left to allocate wisely.")
elif leftover == 0:
    print(" I am broke. No extra room, but i am not overspending.")
else:
    print("Overspending alert, I spent more than I earn.")
```

## Scene 5: The Wallet Wizard’s Rulebook (50/30/20)  
A simple guide:

- **Needs (50%)**: rent, food, transport, bills  
- **Wants (30%)**: Michellaneous  
- **Savings (20%)**: savings  

Now I’ll calculate what my ideal targets should be.


```python
def percent(amount, income):
    return round((amount / income) * 100, 1)

needs = spending["Rent"] + spending["Food"] + spending["Transport"] + spending["Bills"]
wants = spending["michellaneous"]
savings = spending["Savings"]

print("Needs:", needs, f"({percent(needs, monthly_income)}%)")
print("Wants:", wants, f"({percent(wants, monthly_income)}%)")
print("Savings:", savings, f"({percent(savings, monthly_income)}%)")
```

```python
import matplotlib.pyplot as plt

labels = ["Needs", "Wants", "Savings"]
values = [needs, wants, savings]

plt.figure(figsize=(6,4))
plt.bar(labels, values)
plt.title("My Budget Breakdown")
plt.ylabel("Amount ($)")
plt.show()
```
